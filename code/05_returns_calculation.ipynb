```python
#qtr end based on which to form the portfolio, example: Jun-19, Mar-19, Dec-18, Sep-18
#percentile is the long cutoff (number, not a string)
def port_form (qtr_end,percentile):
    long_cutoff = 0
    short_cutoff = 0
    #print (cutoff.iloc[2]['QtrEnd'])
    #print(qtr_end)
    for i in range(len(cutoff)):
        if cutoff.iloc[i]['QtrEnd']==qtr_end:
            long_cutoff = cutoff.iloc[i-1][str(percentile)]
            short_cutoff = cutoff.iloc[i-1][str(100-percentile)]
    long_st=[]
    short_st=[]
    for i in range(len(sue)):
        if sue.iloc[i][qtr_end]>=long_cutoff:
            long_st.append(sue.iloc[i]['Ticker'])
        elif sue.iloc[i][qtr_end]<=short_cutoff:
            short_st.append(sue.iloc[i]['Ticker'])
    return long_st, short_st
    
    
```

```python
def announce_date (ticker,qtr_end):
    ann_date = 0
    qtr_dt = pd.to_datetime(qtr_end)
    # if real stock is entered and if date is really at quarter end, if thats true, announcement date
    if (ticker in ann_dates['Ticker'].values)&(((qtr_dt.day == 30)& (qtr_dt.month in [6,9])) 
                                        or ((qtr_dt.day == 31)& (qtr_dt.month in [3,12]))):
        ann_date = ann_dates.loc[ann_dates['Ticker']==ticker].iloc[0][qtr_end]
#         print(temp)
    else:
        print('not qtr end or not valid stock')
        return None
    return ann_date
```

```python
def trade (ticker,ann_date,SL=-0.05,TP=0.1,ndays=40,posn='long'):
    #posn = position.copy()
    #print (ticker)     
    print(ticker)
    
    names = []

    i =0
    #this is correct, i should put an i = 0 here
    price = mod_prices[mod_prices['Ticker']==ticker]
    price.reset_index(inplace=True)
   
    #before the for loop, i is '2017-06-30'
    
    for i in range (len(price)):
        # once it goes into the for loop, i is 0 and then loop into 1,2,3,...
#         print(i)
        if price.iloc[i]['date']>=ann_date:
            break
            
#     print(f'last 1 i: {i}')

#     print(i) # i should be 912 here, but its 0 i think
    start = price.iloc[i]['date']
#     print(f'last 2 i: {i}')

    #print(start)
    end = price.iloc[i+ int(ndays)]['date']
    #print(end)
    
    price = price[price['date']>=start]
    price = price[price['date']<end]
    
    price.reset_index(inplace=True)
    price.drop('index',axis=1,inplace=True)
#     print(price)
    
#     print(f'last 3 i: {i}')

    
    buy_price = price.iloc[0]['PX_LAST']
    trade_end = ''
    for i in range(1,len(price)):
        if posn == 'long':
            if (price.iloc[i]['PX_LAST']-buy_price)/buy_price<=SL:
                sell_price = price.iloc[i]['PX_LAST']
                posn = 'squared off'
                trade_end = 'SLH'
                profit = (sell_price-buy_price)/buy_price
                #i believe that profit is in percentage
                break
            elif (price.iloc[i]['PX_LAST']-buy_price)/buy_price>=TP:
                sell_price = price.iloc[i]['PX_LAST']
                posn = 'squared off'
                trade_end = 'PB'
                profit = (sell_price-buy_price)/buy_price
                break
        if posn == 'short':
            if (-(price.iloc[i]['PX_LAST']-buy_price))/buy_price<=SL:
                sell_price = price.iloc[i]['PX_LAST']
                posn = 'squared off'
                trade_end = 'SLH'
                profit = (-(sell_price-buy_price))/buy_price
                break
            elif (-(price.iloc[i]['PX_LAST']-buy_price))/buy_price>=TP:
                sell_price = price.iloc[i]['PX_LAST']
                posn = 'squared off'
                trade_end = 'PB'
                profit = (-(sell_price-buy_price))/buy_price
                break
    
        if posn!='squared off':
            sell_price = price.iloc[len(price)-1]['PX_LAST']
            trade_end = 'PSO'
            if posn=='long':
                profit = (sell_price-buy_price)/buy_price
            else:
                profit = (-(sell_price-buy_price))/buy_price
    
#     print(f'last 4 i: {i}')
#     print(ticker)
                    
    return buy_price, sell_price, profit, trade_end
```
